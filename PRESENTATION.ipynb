## Introduction 

Pour comprendre la théorie du random forest il faut bien comprendre : 
- Les arbres de décision
- L’ensemble  learning

### Les arbres de décision

Les arbres de décision sont des algorithmes d'apprentissage automatique 
polyvalent qui peuvent effectuer des tâches de classification et de régression. 
Ce sont des algorithmes très puissants, capables de s'adapter à des ensembles 
de données complexes.

### L'ensemble learning
De maniere plus simple l'ensemble learning cest le fait d'entrainer petits models de Machines learning avec peu de performances pour creer un gros Models avec de trés bonnes performances.
Pour cela On utilise destechniques d'apprentissages de ML par exemple le Bagging;
<br> Ce principe se base:
- L'idée de creer plusieurs copies d'un model en entrainant chaque copie sur une part$ie aleatoire du dataset.
Pour cet algorithme on utilise une technique d'echantillonnage appelée le *Bootstraping*
- Bootstraping consiste à remplacer aprés tirage au sort les données qui ont été selectionnées dans notre dataset

### La sagesse de la foule<br>
C’est une théorie qui présuppose que la perception et la résolution d'un problème sont plus efficaces par une foule que par n'importe quel individu en faisant partie ou non.
Selon ce concept, une foule d’amateur peut mieux répondre à un problème d’un seul expert. 
<br>La foule cependant doit répondre à trois hypothèses pour valider cette théorie :
- la diversité : avoir des personnes de divers milieux avec des idées originales ;
- l’indépendance : permettre à ces avis divers de s’exprimer sans aucune influence ;
- la décentralisation : laisser ces différents jugements s’additionner plutôt que de laisser une autorité supérieure choisir les idées qu’elle préfère.

### QUAND UTILISER UN RANDOM FOREST ? 
Random Forest est envisageable pour traiter les problématiques de prédiction. 
Il est possible de s’en servir dans des challenges Data Science, avec l’assurance d’obtenir de bons résultats. 
Les forêts aléatoires sont exploitables quand on a un nombre important de variables explicatives. 
Cette méthode s’appuie sur des techniques novatrices permettant de classer les prédicteurs par ordre d’importance.

### AVANTAGE DU RANDOM FOREST
La méthode du Random Forest permet de présenter des estimations dont l’importance est variable, c’est-à-dire des réseaux neuronaux.<br>
Son utilisation favorise un travail avec des données manquantes, 
car celles-ci sont remplacées par la variable apparaissant dans un nœud particulier.

### GESTION DES PARAMÈTRES
Avec le Random Forest, il n’y a pas de sur-apprentissage. <br>
Il est ainsi possible de réaliser une parallélisation et un calibrage plus aisé des paramètres. <br>
Contrairement à d’autres techniques, il n’y a que très peu d’hyperparamètres à gérer

Le Random Forest traite aussi bien les problèmes de régression et de classification
Nous allons prendre un exemple de classifications avec le dataset d'iris

```python
##Importation 
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

##Données

from sklearn.datasets import load_iris


iris = load_iris()
x=iris.data
y=iris.target
# print(iris)
iris.feature_names
for i in iris:
      print(i)
print(iris.target_names)
print(sns.pairplot(data=pd.DataFrame(x)))
# y=df['survived']
# x=df.drop(['survived'], axis=1)
```

```python
print(iris.feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=12)
```

```python
model=RandomForestClassifier(criterion='gini',n_estimators=100,max_depth=5,min_samples_leaf=10,random_state=123)
```

```python
model.fit(X_train,y_train)
for i in model.estimators_[:3]:
    tree.plot_tree(i)
    plt.show()
    print(i.score(X_test,y_test))
```

```python
y_pred=model.predict(X_test)
```

```python
# print(np.sum(y_pred == y_test)/X_test.shape[0])
```

```python
model.score(X_test,y_test)
```
